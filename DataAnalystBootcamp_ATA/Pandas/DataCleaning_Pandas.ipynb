```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Herenya\Desktop\Road to Data Analyst\DataAnalystBootcampATA\Pandas\Customer Call List.xlsx")

df
```

```python
#Start cleaning: drop duplicates

clean_df = df.drop_duplicates()
```

```python
clean_df
```

```python
#Remove info you don't need / aren't going to use

clean_df = clean_df.drop(columns = 'Not_Useful_Column')

```

```python
clean_df['Last_Name'] = clean_df["Last_Name"].str.lstrip('...')

clean_df['Last_Name'] = clean_df["Last_Name"].str.lstrip('/')

clean_df['Last_Name'] = clean_df["Last_Name"].str.rstrip('_')

#also: clean_df["Last_Name"].str.strip('./_') ---> Combine everything you uwant to remove in a single string, no lists
```

```python
#Cleaning and standarizing phone number: first clean, leave all numbers, then format how you want

#We're using a regex saying: match every character that is NOT: a letter (lower or uppercase) or a number and replace it with nothing ' ' (empty space)

clean_df['Phone_Number'] = clean_df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex = True)

```

```python

#Format using lambda: we want three character, slash, three characters, slash, three ch. We need to convert al phone num to string for it to work
#Can also be done with for loops, but lamda is quicker
clean_df['Phone_Number'] = clean_df['Phone_Number'].apply(lambda x: str(x))
clean_df['Phone_Number'] = clean_df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
clean_df["Phone_Number"] = clean_df["Phone_Number"].str.replace('nan--','')
clean_df["Phone_Number"] = clean_df["Phone_Number"].str.replace('Na--','')
```

```python
clean_df[["Street_Address", "State", "Zip_Code"]] = clean_df["Address"].str.split(',', n=2, expand = True)
```

```python
#Standarize Y/N yes/no columns:
#clean_df['Do_Not_Contact'].str.replace('[^Yesy', 'Yes')

clean_df['Do_Not_Contact'] = clean_df['Do_Not_Contact'].str.replace('Yes', 'Y')

clean_df['Do_Not_Contact'] = clean_df['Do_Not_Contact'].str.replace('No', 'N')

```

```python
clean_df['Paying Customer'] = clean_df['Paying Customer'].str.replace('Yes', 'Y')

clean_df['Paying Customer'] = clean_df['Paying Customer'].str.replace('No', 'N')
```

```python
clean_df = clean_df.replace('N/a','')
#clean_df.replace('NaN','')
#Doesn't work with Nan because those cells are empty, there is nothing in there, it just shows NaN. So we have to fill it with ' '  blank

clean_df = clean_df.fillna('')
clean_df
```

```python
#Now we want to remove people without phone numbers or people who we can't contact
#We use the index to loop through and then indicate to look only at the value we're interested in

for x in clean_df.index:
    if clean_df.loc[x, "Do_Not_Contact"] == 'Y':
        clean_df.drop(x, inplace=True)
```

```python
for x in clean_df.index:
    if clean_df.loc[x, "Phone_Number"] == '':
        clean_df.drop(x, inplace=True)

#Another way to drop null values: dropna
#clean_df = clean_df.dropna(subset="Phone_Number"), inplace=True)
```

```python
clean_df
```

```python
#Since we deleted rows we can reset the index for it to be made of consecutive numbers, and less confusing

clean_df = clean_df.reset_index(drop=True)

#We include the drop = true because otherwise it keeps the old index in a new column we don't usually need
```

```python
#reminder: types of join
# inner join: only takes what is present in both datasets (only matches) Present in both left and right dataframes
# outer join: takes everything from the two datasets (everything from left, right and what's similar)
# left or right join: takes everything from one (left or right) and then only what is similar (matches) from the other dataframe.
#For example left join takes everything from left, what's similar to left from right, but NOT WHAT'S UNIQUE TO RIGHT.
```

```python
df1.merge(df2)

#default is inner join, you can specify otherwhise
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'cross')

#compares each value of one dataset to each value of the other
```

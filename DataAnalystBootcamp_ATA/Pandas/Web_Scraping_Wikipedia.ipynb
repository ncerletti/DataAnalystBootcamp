```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table')
```

```python
soup.find('table', class_ = 'wikitable sortable')

#find is going to pull out the first one of that class
```

```python
#another way, find_all and indexing:

table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
body_info = table.find_all('td')
print(body_info)
```

```python
import re

count_i = 1

col_rank = []

col_company = []

col_industry = []

col_revenue = []

col_revgrowth = []

col_employees = []

col_hq = []


for info in body_info:
    if count_i == 1:
        col_rank.append((info.text).strip())
        count_i = count_i + 1
    elif count_i == 2:
        col_company.append((info.text).strip())
        count_i = count_i + 1
    elif count_i == 3:
        col_industry.append((info.text).strip())
        count_i = count_i + 1
    elif count_i == 4:
        col_revenue.append((info.text).strip())
        count_i = count_i + 1
    elif count_i == 5:
        col_revgrowth.append((info.text).strip())
        count_i = count_i + 1
    elif count_i == 6:
        col_employees.append((info.text).strip())
        count_i = count_i + 1
    elif count_i == 7:
        col_hq.append((info.text).strip())
        count_i = 1
```

```python
print(col_rank)
print(len(col_rank))

print(col_company)
print(len(col_company))

print(col_industry)
print(len(col_industry))

print(col_revenue)
print(len(col_revenue))

print(col_revgrowth)
print(len(col_revgrowth))

print(col_employees)
print(len(col_employees))

print(col_hq)
print(len(col_hq))
```

```python
table.find_all('th')
table_columns = []
for column in table.find_all('th'):
    table_columns.append((column.text).strip())
print(table_columns)
```

```python
for column in table_columns:
    print(column, end=' | ')
```

```python
for column in table_columns:
    print(column, end=' | ')

print('\n')

for i in range(len(col_rank)):
    print(col_rank[i] + ' | ' +  col_company[i] + ' | ' + col_industry[i] + ' | ' + col_revenue[i] + ' | ' + col_revgrowth[i] + ' | ' + col_employees[i] + ' | ' + col_hq[i])
```
